```python
df = pd.read_csv('/kaggle/input/e4-e7-reads/reads_lengths.txt', sep=' ', header=None)
df
```

```python
df = df.iloc[:, [6, 7]]  # Selecting the 6th and 7th columns, as Python uses 0-based indexing
df
```

```python
# Replace 'old_name_1' and 'old_name_2' with the current names of the 6th and 7th columns
df = df.rename(columns={6: 'Count of Reads', 7: 'Read length'})
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Histogram of the Read Lengths
plt.figure(figsize=(15, 10))
plt.hist(df['Read length'], bins=50, color='skyblue', edgecolor='black')
plt.title('Histogram of Read Lengths')
plt.xlabel('Read Length')
plt.ylabel('Count of Reads')
plt.grid(True)
plt.savefig('histogram_read_lengths.png')  # Saves the plot as a PNG file
plt.show()

# Scatter plot of Count of Reads vs Read Length
plt.figure(figsize=(15, 10))
plt.scatter(df['Read length'], df['Count of Reads'], alpha=0.5, color='tomato')
plt.title('Scatter Plot of Count of Reads vs Read Length')
plt.xlabel('Read Length')
plt.ylabel('Count of Reads')
plt.grid(True)
plt.savefig('scatterplot_reads_vs_length.png')  # Saves the plot as a PNG file
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
# df = pd.read_csv('your_data.csv')  # Uncomment and replace with your actual file path

# Histogram of the Read Lengths
plt.figure(figsize=(15, 10))
n, bins, patches = plt.hist(df['Read length'], bins=50, color='skyblue', edgecolor='black')

# Define a fixed offset for the annotation to be above the bar
fixed_offset = max(n) * 0.01  # 1% of the max height of the bars for spacing

# Extend the upper limit of y-axis to make room for annotation
plt.ylim(top=plt.ylim()[1] + fixed_offset * 15)  # Increase the top y-limit to make room for annotations

# Annotate the range of each bin and the count of reads on the bars vertically
for count, bin_left, bin_right, patch in zip(n, bins, bins[1:], patches):
    bin_range = f"{int(bin_left)}-{int(bin_right)}"
    annotation_text = f"{bin_range} > {int(count)}"
    # Use the fixed offset to position the annotation
    y_value = patch.get_height() + fixed_offset
    plt.text(patch.get_x() + patch.get_width() / 2, y_value, annotation_text,
             ha='center', va='bottom', fontsize=6, rotation=90)  # Reduce font size to fit

plt.title('Histogram of Read Lengths (E4_WT_E14.5_cerebrum)')
plt.xlabel('Read Length')
plt.ylabel('Count of Reads')
# The plt.grid() function is omitted to remove the grid lines
plt.savefig('histogram_read_lengths_E4.pdf', format='pdf')  # Saves the plot as a PDF file
plt.show()

# Scatter plot of Count of Reads vs Read Length
plt.figure(figsize=(15, 10))
plt.scatter(df['Read length'], df['Count of Reads'], alpha=0.5, color='tomato')
plt.title('Scatter Plot of Count of Reads vs Read Length (E4_WT_E14.5_cerebrum)')
plt.xlabel('Read Length')
plt.ylabel('Count of Reads')
plt.grid(True)
plt.savefig('scatterplot_reads_vs_length_E4.pdf', format='pdf')  # Saves the plot as a PDF file
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Histogram of the Read Lengths with logarithmic bins
plt.figure(figsize=(15, 10))
# Create logarithmically spaced bins
bins = np.logspace(np.log10(df['Read length'].min()), np.log10(df['Read length'].max()), 100)
plt.hist(df['Read length'], bins=bins, color='skyblue', edgecolor='black')
plt.xscale('log')  # Set x-axis to logarithmic scale
plt.title('Histogram of Read Lengths')
plt.xlabel('Read Length (log scale)')
plt.ylabel('Count of Reads')
plt.grid(True, which="both", ls="-")  # Grid lines on log scale
plt.savefig('histogram_read_lengths_log_scale.png')
plt.show()

# Scatter plot of Count of Reads vs Read Length with a logarithmic scale
plt.figure(figsize=(15, 10))
plt.scatter(df['Read length'], df['Count of Reads'], alpha=0.5, color='tomato')
plt.xscale('log')  # Set x-axis to logarithmic scale
plt.yscale('log')  # Set y-axis to logarithmic scale if counts are also large
plt.title('Scatter Plot of Count of Reads vs Read Length (log-log scale)')
plt.xlabel('Read Length (log scale)')
plt.ylabel('Count of Reads (log scale)')
plt.grid(True, which="both", ls="-")  # Grid lines on log scale
plt.savefig('scatterplot_reads_vs_length_log_scale.png')
plt.show()
```
